```python
%%writefile punto1.py
import streamlit as st
import matplotlib.pyplot as plt
import numpy as np

st.set_page_config(page_title="Espectro de Fourier")

st.markdown("# Espectro de Fourier")
st.sidebar.header("Espectro de Fourier")
st.write(
    """Este punto muestra el procedimiento y las graficas respectivas de interes para analizar el
    espectro de Fourier, mostrando su magnitud, fase, error relativa y una gr√°fica interactiva donde
    se puede observar tambien la se√±al recosntruida desde 1 hasta 50 arm√≥nicos. Disfrutalo!"""
)


st.markdown("## Planteamiento inicial")
st.write(
    """Encuentre la expresi√≥n del espectro de Fourier (forma exponencial y trigonom√©trica) para la se√±al
    $x(t) = \lvert A \sin(2 \pi F_0 t) \\rvert ^2 $, con $t \in [- \\frac{1}{2 F_0}, \\frac{1}{2 F_0}]$, con
    $A, F_0 \in \mathbb{R}^+$"""
)


# Par√°metros
A = 1
Fo = 100

# C√°lculo de tiempos y frecuencias
ti = -1 / (2 * Fo)  # tiempo inicial
tf = 1 / (2 * Fo)   # tiempo final
T = tf - ti
wo = 2 * np.pi / T  # frecuencia angular
Fo = wo / (2 * np.pi)  # frecuencia fundamental

# Frecuencia de muestreo
Fs = 30 * Fo  # frecuencia de muestreo -> debe respetar Nyquist Fs >= 2 * Fo
tv = np.arange(ti, tf, 1 / Fs)  # vector de tiempo
Nm = len(tv)  # n√∫mero de muestras en el tiempo

# Se√±al de entrada
xo = (abs(A * np.sin(wo * tv))) ** 2  # se√±al de entrada

# Crear la gr√°fica
fig, ax = plt.subplots()
ax.plot(tv, xo)
ax.set_title('Gr√°fica de la Se√±al de Entrada')
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('Amplitud')

# Mostrar la gr√°fica en Streamlit
st.pyplot(fig)

###################
st.markdown("## Gr√°ficas Magnitud y Fase en dB")

# Explicaci√≥n inicial
st.write(r"""
Dado que la se√±al presenta simetr√≠a **PAR**, es decir $x(t) = x(-t)$, su espectro de Fourier solo tendr√° componente $\mathbb{R}$. De los c√°lculos presentados en el PDF, se tiene que:
""")

# Mostrar la primera ecuaci√≥n en un recuadro
st.latex(r"\boxed{b_n = 0 \quad \forall \ n \in \{1,\dots,N\}}")

# Continuaci√≥n de la explicaci√≥n
st.write("Por lo tanto, el espectro de Fourier en su forma trigonom√©trica es:")

# Mostrar la segunda ecuaci√≥n en un recuadro
st.latex(r"""
\boxed{a_n = \left\{\begin{align}
0 & \quad \forall \ n \setminus \{0, 2\}\\
\frac{A^2}{2} & \quad n = 0\\
-\frac{A^2}{2} & \quad n = 2\\
\end{align}\right.}
""")

# Mostrar la tercera ecuaci√≥n en un recuadro
st.latex(r"""
\boxed{x(t)= a_0 + \sum_{n=1}^N{a_n\cos(n\omega_0 t)} = a_0 + a_2 \cos(n\omega_0 t) =  \frac{A^2}{2} - \frac{A^2}{2} \cos(2 \cdot 2\pi F_0 t)}
""")

# Explicaci√≥n sobre la forma exponencial compleja
st.write("Para el caso de la exponencial compleja se tiene que:")

# Mostrar la cuarta ecuaci√≥n
st.latex(r"c_0 = a_0 = \frac{A^2}{2}")

st.write(r"""
Y haciendo uso de la igualdad entre los coeficientes \(c_n \) y \( a_n, \ b_n\), se obtiene lo siguiente:
""")

# Mostrar la quinta ecuaci√≥n
st.latex(r"c_n = \frac{a_n - j b_n}{2}")

st.write("Por lo tanto:")

# Mostrar la sexta ecuaci√≥n en un recuadro
st.latex(r"""
\boxed{c_n = \left\{\begin{align}
- \frac{A^2}{4} & \quad n = -2\\
\frac{A^2}{2} & \quad n = 0\\
-\frac{A^2}{4} & \quad n = 2\\
0 & \quad \forall \ n \setminus \{-2,0,2\}
\end{align}\right.}
""")

# Explicaci√≥n sobre el espectro en forma exponencial compleja
st.write(r'Por lo tanto el espectro de Fourier en su forma exponencial compleja es:')

st.latex(r'\boxed{x(t) = \sum_{n=-N}^{N}{c_n e^{j n \omega_0 t}}}')

st.latex(r'x(t) = c_{-2}e^{-j2 \cdot 2\pi F_0t} + c_{0}e^{0} + c_{2}e^{j2\cdot 2\pi F_0t}')

st.latex(r'\boxed{x(t) = -\frac{A^2}{4} [\cos(2 \cdot 2\pi F_0 t) - j \sin(2 \cdot 2\pi F_0 t)] + \frac{A^2}{2} -\frac{A^2}{4} [\cos(2 \cdot 2\pi F_0 t) + j \sin(2 \cdot 2\pi F_0 t)]}')


st.write(r'Teniendo las series definidas se procede a realizar las simulaciones respectivas para mostrar su magnitud, fase, error relativo y la se√±al reconstruida para $N = \{ 1,2,...,50 \}$')

#se presenta la simulaci√≥n del espectro y la reconstrucci√≥n
#construir bases serie exponencial
N = 50 #numero armonicos
wo = 2*np.pi/T # frecuencia fundamental

# definir bases
phin = np.zeros((len(tv),2*N+1),dtype=np.complex_)#crear matriz para guardar bases
for n in range(-N,N+1,1):
    phin[:,n+N] = np.exp(1j*n*wo*tv)#base de fourier en el intervalo de inter√©s

#espectro de la se√±al - serie exponencial
#calcular espectro -> cambiar seg√∫n la se√±al estudiada
Cn = np.zeros(2*N+1,dtype=np.complex_)
nv = np.linspace(-N, N, 2*N+1) #vector num arm√≥nicos
Cn[N] = (A**2) / 2 # nivel dc C_0
Cn[N-2] = -(A**2) /4
Cn[N+2] = -(A**2) /4


#graficar espectro
fig = plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.stem(nv, np.real(Cn), 'r')
plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
plt.ylabel(r'$Re\{C_n\}$',fontsize = 14)
plt.axis('tight')
plt.grid()

plt.subplot(2,2,2)
plt.stem(nv, np.imag(Cn), 'r')
plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
plt.ylabel(r'$Im\{C_n\}$',fontsize = 14)
plt.axis('tight')
plt.grid()

plt.subplot(2,2,3)
plt.stem(nv, 20*np.log10(abs(Cn)), 'r')#magnitud del espectro
plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
plt.ylabel(r'$|C_n| \ [dB]$',fontsize = 14)
plt.axis('tight')
plt.grid()

plt.subplot(2,2,4)
plt.stem(nv, np.angle(Cn), 'r')#fase del espectro
plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
plt.ylabel(r'$\langle C_n$',fontsize = 14)
plt.axis('tight')
plt.grid()

fig.tight_layout()

# Mostrar la figura en Streamlit
st.pyplot(fig)

st.title('Reconstrucci√≥n de la se√±al y Error relativo')

# Reconstructing
Px = (3 * (A**4)) / 8

# Streamlit Slider
n = st.slider("Selecciona el n√∫mero de arm√≥nicos (n)", 0, N, 1)

# Crear una figura para Streamlit
fig, ax = plt.subplots(2, 3, figsize=(12, 5))

# Funci√≥n para actualizar gr√°ficos
def update(n):
    # Limpiar las subplots
    ax[0, 0].clear()
    ax[0, 1].clear()
    ax[0, 2].clear()
    ax[1, 0].clear()
    ax[1, 1].clear()
    ax[1, 2].clear()

    # Espectro
    ind = np.arange(N - n, N + n + 1)  # Arm√≥nicos a considerar
    ax[0, 0].stem(nv, abs(Cn), 'b')  # Magnitud del espectro
    ax[0, 0].stem(nv[ind], abs(Cn[ind]), 'r')  # Magnitud del espectro
    ax[0, 0].set_xlabel(r'$nw_o$[rad/s]')
    ax[0, 0].set_ylabel(r'$|C_n|$')

    # Componentes espectrales
    ce = (phin[:, ind]).dot(np.diag(Cn[ind]))
    ax[0, 1].plot(np.real(ce))  # Componente real
    ax[0, 1].set_xlabel('t[s]')
    ax[0, 1].set_ylabel('$Re\{c_n\phi_n(t)\}$')

    ax[0, 2].plot(np.imag(ce))  # Componente imaginario
    ax[0, 2].set_xlabel('t[s]')
    ax[0, 2].set_ylabel('$Im\{c_n\phi_n(t)\}$')

    # Reconstrucci√≥n
    er = 1 - ((np.sum(abs(Cn[ind])**2)) / Px)
    # Se√±al reconstruida
    xe = (phin[:, ind]).dot(Cn[ind])
    # Error en t
    et = xo - xe

    ax[1, 1].plot(tv, xo, 'r', linewidth=4, label='$x(t)$')
    ax[1, 1].plot(tv, xe, color='b', label='$x_e(t)$')  # Se√±al estimada o filtrada
    ax[1, 1].plot(tv, et, color='g', label='$e(t)$')  # Se√±al estimada o filtrada
    ax[1, 1].set_title(f'$n={n}. E_r$={100*er:.2f}[%%]')
    ax[1, 1].set_xlabel('t[s]')
    ax[1, 1].set_ylabel('x(t)')
    ax[1, 1].legend(loc='center left', bbox_to_anchor=(1, 0.5))
    ax[1, 0].axis('off')
    ax[1, 2].axis('off')
    plt.subplots_adjust(wspace=0.5, hspace=0.7)

# Actualizar la figura con el valor del slider
update(n)

# Mostrar la figura en Streamlit
st.pyplot(fig)
```

```python
%%writefile punto3.py

import streamlit as st
import matplotlib.pyplot as plt
import numpy as np
import soundfile as sf # para instalar pip install soundfile
import subprocess
import os

st.set_page_config(page_title="Modulaci√≥n AM", page_icon="üìª")

st.markdown("# Modulaci√≥n AM")
st.sidebar.header("Modulaci√≥n AM")
st.write(
    """Este punto muestra el procedimiento para la modulaci√≥n en amplitud (AM) de un mensaje, en este
    caso cualquier canci√≥n de YouTube en la cual solo se procesan 5 segundos (desde el segundo 20 al 25),
    el usuario puede ingresar cualquier valor para el indice de modulaci√≥n, asi como ingresar cualquier link de
    YouTube para procesar."""
)




# Funci√≥n para descargar el audio usando yt-dlp
def download_audio(link):
    try:
        # Ejecutar el comando yt-dlp para descargar el audio
        result = subprocess.run(
            ["yt-dlp", "--extract-audio", "--audio-format", "mp3", "-o", "audio.mp3", "--force-overwrites", link],
            capture_output=True, text=True
        )
        if result.returncode == 0:
            st.success("Audio descargado exitosamente.")
        else:
            st.error(f"Error al descargar el audio: {result.stderr}")
    except Exception as e:
        st.error(f"Error al ejecutar el comando: {e}")

# Funci√≥n para convertir MP3 a WAV usando ffmpeg
def convert_to_wav():
    try:
        # Ejecutar el comando ffmpeg para convertir MP3 a WAV
        result = subprocess.run(
            ["ffmpeg", "-y", "-i", "audio.mp3", "output.wav"],
            capture_output=True, text=True
        )
        if result.returncode == 0:
            st.success("Conversi√≥n a WAV completada.")
        else:
            st.error(f"Error al convertir el archivo: {result.stderr}")
    except Exception as e:
        st.error(f"Error al ejecutar el comando: {e}")



# Interfaz en Streamlit
st.markdown("## Procesar Audio de YouTube")

# Enlace de YouTube ingresado por el usuario
link = st.text_input("Introduce el enlace de YouTube:", "https://www.youtube.com/watch?v=DU5fiwedSCk")

# Si el enlace es proporcionado, ejecutar las funciones
if link:
    download_audio(link)
    convert_to_wav()


import numpy as np
import soundfile as sf

#lee archivos wav
nombre_out = "output.wav"
x, fs = sf.read(nombre_out)

if x.ndim > 1:
    x = np.mean(x, axis=1)

# Extraer los 5 segundos desde los 20 a los 25 segundos
start_sample = int(20 * fs)
end_sample = int(25 * fs)
x_fragment = x[start_sample:end_sample]

#Reproducir el fragmento seleccionado
with open('output.wav', 'rb') as audio_file:
    st.audio(audio_file, format='audio/wav', start_time=20, end_time=25)

import matplotlib.pyplot as plt

st.markdown("## Gr√°ficas en tiempo y frecuencia")

#par√°metros de la portadora
Fc = 10000  #frecuencia de la portadora
Imx = st.number_input("Ingrese el indice de modulaci√≥n:", 0.8 )
Im = Imx #Im = peak(m(t))/Ac ->  Ac = peak(m(t))/Im
Fs = 5 * Fc

t = np.arange(len(x_fragment)) / fs #vector del tiempo

m = x_fragment / np.max(np.abs(x_fragment)) #normalizar el fragmento de audio -> Mensaje

Ac = np.max(np.abs(m)) / Im #amplitud de la portadora seg√∫n √≠ndice de modulaci√≥n
c = Ac * np.sin(2 * np.pi * Fc * t)  #portadora

print(Ac)

st.markdown("### Portadora y mensaje")
# Graficar la se√±al portadora y el mensaje
fig = plt.figure(figsize=(10, 6))
plt.plot(t, c, label='Portadora')
plt.plot(t, m, label='Mensaje')
plt.legend()
plt.xlabel('Tiempo [s]')

st.pyplot(fig)

st.markdown("### Se√±al modulada")
# Se√±al modulada en AM
y = (1 + m / Ac) * c

# Graficar la se√±al modulada
fig = plt.figure(figsize=(10, 6))
plt.plot(t, y)
plt.xlabel('Tiempo [s]')
plt.ylabel('Se√±al Modulada AM')
plt.title('Modulaci√≥n AM')

st.pyplot(fig)

st.markdown("### Espectro mensaje, portadora y modulada")
st.write("La se√±al de la portadora se defini√≥ en 10 KHz para el desarrollo de este punto.")
Xfc = np.fft.fft(c)
Xfm = np.fft.fft(m)
Xfy = np.fft.fft(y)

vfre = np.fft.fftfreq(len(c),1/Fs)

fig, ax = plt.subplots()
ax.plot(vfre,abs(Xfy),label='Modulado')
ax.plot(vfre,abs(Xfm),label='Mensaje')
ax.plot(vfre,abs(Xfc),label='Portadora')
ax.legend()
ax.set_xlabel('$f [Hz]$')
ax.set_ylabel('$|X(f)|$')
ax.set_ylim([min(abs(Xfm)), 1.1*max(abs(Xfm))])

st.pyplot(fig)

st.markdown("## Fragmentos de audio")
#Reproducir la se√±al mensaje
st.markdown("### Se√±al mensaje")
st.audio(x_fragment, sample_rate=fs)

#Reproducir la se√±al portadora
st.markdown("### Se√±al portadora")
st.audio(c, sample_rate=fs)

#Reproducir la se√±al modulada
st.markdown("### Se√±al modulada")
st.audio(y, sample_rate=fs)
```

La THD mide la proporci√≥n de la potencia de los arm√≥nicos (componentes de frecuencia m√∫ltiplo de la frecuencia fundamental) respecto a la potencia de la se√±al fundamental. La distorsi√≥n arm√≥nica ocurre cuando una se√±al se altera de manera que se a√±aden frecuencias adicionales (arm√≥nicos) a la se√±al original.


```python
%%writefile punto4.py

import streamlit as st
import matplotlib.pyplot as plt
import numpy as np

st.set_page_config(page_title="THD y DFP", page_icon="‚ö°")

st.markdown("# Total Harmonic Distortion y DFP")
st.sidebar.header("THD y DFP")
st.write(
    """Este punto genera un ejemplo ilustrativo del c√°lculo del THD y DFP para un rectificador de onda completa con
    carga netamente resistiva y una carga RC serie, el usuario puede escoger los valores de R y C respectivamente"""
)

st.header("1. ¬øEn qu√© consiste la distorsi√≥n total de arm√≥nicos (THD)?")
st.write("""
La distorsi√≥n arm√≥nica total se define como la suma de las contribuciones a la distorsi√≥n de cada arm√≥nico.
Sin embargo, no se puede definir como una suma algebraica directa, debido a que los coeficientes de la serie
de Fourier pueden ser positivos o negativos.
""")

st.header("2. ¬øEn qu√© consiste el factor de potencia de un circuito el√©ctrico?")
st.write("""
El factor de potencia (FP) es una medida que indica qu√© tan eficientemente un circuito el√©ctrico utiliza la energ√≠a.
Se define como el cociente entre la potencia activa (real), que realiza trabajo √∫til, y la potencia aparente, que es
la combinaci√≥n de la potencia activa y la reactiva (la que circula entre los elementos reactivos sin realizar trabajo).

Matem√°ticamente, se expresa como:

$$FP = \\cos(\\theta)$$

donde $\\theta$ es el √°ngulo de desfase entre la tensi√≥n y la corriente. Un factor de potencia cercano a 1 significa que
la energ√≠a se usa de manera eficiente, mientras que un valor bajo indica la presencia de potencia reactiva, que no
contribuye al trabajo √∫til pero aumenta las p√©rdidas en el sistema.
""")

st.header("3. ¬øC√≥mo puede calcularse el THD desde la FFT?")
st.write("""
Primero, se debe calcular la FFT de la se√±al y obtener las magnitudes (o valores RMS) correspondientes a las diferentes frecuencias.
Luego, se deben identificar $V_1$, que corresponde a la magnitud de la frecuencia fundamental, y los $V_2, V_3, \\dots, V_n$,
que corresponden a las magnitudes de los arm√≥nicos superiores, estos arm√≥nicos son m√∫ltiplos enteros de la frecuencia fundamental.

Con los valores de frecuencia en Hz y el espectro calculado como arreglo de numpy, se puede calcular la distorsi√≥n total de arm√≥nicos (THD) seg√∫n:

$$THD = \\frac{\\sqrt{V_2^2 + V_3^2 + \\cdots + V_n^2}}{V_1}$$
""")

st.header("4. ¬øC√≥mo puede calcularse la distorsi√≥n del factor de potencia con base al THD?")
st.write("""
La medida de distorsi√≥n del factor de potencia se puede calcular a partir del THD, usando la siguiente f√≥rmula:

$$PF_{THD} = \\sqrt{\\frac{1}{1 + (THD)^2}}$$
""")


st.markdown("# 1. Carga netamente resistiva")
#paquetes de para simulaci√≥n
import numpy as np
import scipy
import matplotlib.pyplot as plt
#%matplotlib inline
import scipy.signal as sig
import scipy.optimize as opt
#from IPython.display import Image
import sympy as sym
#sym.init_session()

s = sym.symbols('s', complex=True)
t, R, L, C = sym.symbols('t R L C', positive=True)
X = sym.Function('X')(s)

Y = 1/(R*C*s + 1) * X


s = sym.symbols('s', complex=True)
t, R, L, C = sym.symbols('t R L C', positive=True)
X = sym.Function('X')(s)

Y = X


R_v = st.number_input("Ingrese el valor de R:", 1000) #valores de simulacion de R

Y_RC = Y.subs(R, R_v)


num = np.array([1]) #coeficientes del numerador de la funcion de transferencia como numpy array
den = np.array([1])#coeficientes del denominador de la funcion de transferencia como numpy array
G_n = sig.TransferFunction(num, den) #funcion de transferencia en
print(num)
print(den)

Fo = 60 #frec alimentaci√≥n
Fs = 30*Fo #frecuencia muestreo
To = 1/Fo #periodo fundamental
Ts = 1/Fs #periodo muestreo
t = np.arange(0, 5*To,Ts) # se simulan 5 peridos de alimentaci√≥n con un muestreo segun Ts
A = 120 #amplitud entrada
#funci√≥n de entrada aproximada como onda senoidal rectificada completa
in_o = abs(A*(np.sin(2*np.pi*Fo*t)))
rec_c = sig.square(2*np.pi*Fo*t)#funci√≥n tren de pulsos para simular se√±al rectificada
#Para rectificador de media onda:
rec_m = 0.5*(sig.square(2*np.pi*Fo*t)+1)
in_ = in_o * rec_c #si se quiere probar con rec media onda multiplicar por rec_m
out = G_n.output(in_o, T=t)[1] #evaluar salida ante entrada rectificada mediante aproximacion num√©rica de scipy

st.markdown("## Gr√°fica se√±al de alimentaci√≥n")
#se√±ales de alimentaci√≥n y tren de pulsos
fig, ax = plt.subplots()
ax.plot(t,in_o)
ax.set_title('Alimentaci√≥n')
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('Amplitud')

st.pyplot(fig)

st.markdown("## Gr√°fica entrada y salida de circuito")
fig, ax = plt.subplots()
ax.plot(t, in_o,label='In(t)')#entrada rectificada
ax.plot(t, out,label='Out(t)')#salida del circuito
ax.set_xlabel('t [s]')
ax.set_ylabel('out(t)')

st.pyplot(fig)

vfre = np.fft.rfftfreq(len(out),1/Fs) #vector de frecuencia en Hz segun tama√±o del vector y tiempo de muestreo
Xf = np.fft.rfft(out) #Fourier de la salida

st.markdown("## C√°lculo del espectro con FFT")
fig, ax = plt.subplots()
ax.stem(vfre,abs(Xf/len(out)))
ax.set_xlabel('Frecuencia [Hz]')
ax.set_ylabel('$|X(f)|$')

st.pyplot(fig)

V1 = np.abs(Xf[1])  # Componente fundamental (frecuencia 60 Hz)
Vn = np.sqrt(np.sum(np.abs(Xf[2:])**2))  # Suma de los arm√≥nicos superiores
THD = Vn / V1  # C√°lculo del THD

DFP = np.sqrt((1/(1+(THD**2))))

st.markdown("## C√°lculo del THD en base al FFT")
# Mostrar el resultado del THD
st.write(f'THD: {THD * 100:.2f}%')

st.markdown("## C√°lculo del DFP en base al THD")
st.write('El DFP es: ', DFP)

st.markdown("## Conclusi√≥n")
st.write(
    """
Cuando una se√±al tiene componentes no senoidales (como la rectificaci√≥n de la se√±al),
se generan arm√≥nicos adicionales que no est√°n presentes en una onda senoidal pura. Esto es lo que
est√° aumentando el THD, no el comportamiento de la carga resistiva.
"""
)



st.markdown("# 2. Carga RC Serie")
#Carga RC serie
s = sym.symbols('s', complex=True)
t, R, L, C = sym.symbols('t R L C', positive=True)
X = sym.Function('X')(s)

Y = 1/(R*C*s + 1) * X


R_v = st.number_input("Ingrese el valor de R:", value=1000, key="R_value") #valores de simulacion de R
C_v = st.number_input("Ingrese el valor de C (Ejemplo 10e-6): ", value=10e-6, key="C_value") # valores de simulacion de C
Y_RC = Y.subs(R, R_v).subs(C, C_v)

num = np.array([1]) #coeficientes del numerador de la funcion de transferencia como numpy array
den = np.array([R_v*C_v,1])#coeficientes del denominador de la funcion de transferencia como numpy array
G_n = sig.TransferFunction(num, den) #funcion de transferencia en
print(num)
print(den)

Fo = 60 #frec alimentaci√≥n
Fs = 30*Fo #frecuencia muestreo
To = 1/Fo #periodo fundamental
Ts = 1/Fs #periodo muestreo
t = np.arange(0, 5*To,Ts) # se simulan 5 peridos de alimentaci√≥n con un muestreo segun Ts
A = 120 #amplitud entrada
#funci√≥n de entrada aproximada como onda senoidal rectificada completa
in_o = abs(A*(np.sin(2*np.pi*Fo*t)))
rec_c = sig.square(2*np.pi*Fo*t)#funci√≥n tren de pulsos para simular se√±al rectificada
#Para rectificador de media onda:
rec_m = 0.5*(sig.square(2*np.pi*Fo*t)+1)
in_ = in_o * rec_c #si se quiere probar con rec media onda multiplicar por rec_m
out = G_n.output(in_o, T=t)[1] #evaluar salida ante entrada rectificada mediante aproximacion num√©rica de scipy

st.markdown("## Gr√°fica se√±al de alimentaci√≥n")
#se√±ales de alimentaci√≥n y tren de pulsos
fig, ax = plt.subplots()
ax.plot(t,in_o)
ax.set_title('Alimentaci√≥n')
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('Amplitud')

st.pyplot(fig)

st.markdown("## Gr√°fica entrada y salida de circuito")
fig, ax = plt.subplots()
ax.plot(t, in_o,label='In(t)')#entrada rectificada
ax.plot(t, out,label='Out(t)')#salida del circuito
ax.set_xlabel('t [s]')
ax.set_ylabel('out(t)')

st.pyplot(fig)

vfre = np.fft.rfftfreq(len(out),1/Fs) #vector de frecuencia en Hz segun tama√±o del vector y tiempo de muestreo
Xf = np.fft.rfft(out) #Fourier de la salida

st.markdown("## C√°lculo del espectro con FFT")
fig, ax = plt.subplots()
ax.stem(vfre,abs(Xf/len(out)))
ax.set_xlabel('Frecuencia [Hz]')
ax.set_ylabel('$|X(f)|$')

st.pyplot(fig)

V1 = np.abs(Xf[1])  # Componente fundamental (frecuencia 60 Hz)
Vn = np.sqrt(np.sum(np.abs(Xf[2:])**2))  # Suma de los arm√≥nicos superiores
THD = Vn / V1  # C√°lculo del THD

st.markdown("## C√°lculo del THD en base al FFT")
# Mostrar el resultado del THD
st.write(f'THD: {THD * 100:.2f}%')

st.markdown("## C√°lculo del DFP en base al THD")
st.write('El DFP es: ', DFP)

st.markdown("## Conclusi√≥n")
st.write(
    """
El hecho de que el THD sea m√°s bajo en el caso de la carga RC en serie, a pesar de
estar alimentando el sistema con la misma se√±al senoidal rectificada, se debe a que el
condensador introduce un efecto de filtrado en la se√±al.

Un circuito RC en serie act√∫a como un filtro pasa-bajas, lo que significa que aten√∫a las
frecuencias altas (arm√≥nicos superiores) de la se√±al de entrada. En el caso de una se√±al
senoidal rectificada, los arm√≥nicos generados por la rectificaci√≥n tienden a ser de
frecuencias m√°s altas que la componente fundamental.
"""
)
```
