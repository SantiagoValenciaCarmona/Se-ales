##**Funciones de transferencia**

```python
%%writefile 2_📈_Funciones_de_transferencia.py

import streamlit as st
import matplotlib.pyplot as plt
import numpy as np

st.set_page_config(page_title="Funciones de transferencia", page_icon="📈")

st.markdown("# Funciones de transferencia")
st.sidebar.header("Funciones de transferencia")
st.write(
    """Este punto muestra el procedimiento y las graficas respectivas de interes para analizar las
    funciones de transferencia de un pendulo elástico su equivalente en un circuito RLC serie y sus
    respuestas ante diferentes entradas como el impulso y el escalón, el usuario puede definir el factor
    de amortiguamiento para simular diferentes respuestas del sistema como Subamortiguada, criticamente amortiguada,
     sobreamortiguada, inestable y oscilatorio.
    """
)



st.markdown("## Planteamiento inicial")
st.write(
    """Encuentre la función de transferencia que caracteriza el sistema de masa, resorte, amortiguador,
    presentado en la siguiente figura (asuma condiciones iniciales cero). Luego, encuentre el sistema equivalente
    a partir de un circuito RLC (entrada de tensión - salida tensión del capacitor).
    """
)

st.image("https://github.com/amalvarezme/SenalesSistemas/blob/master/4_Transformada_Laplace/damped_spring.png?raw=true")

###################
st.markdown("## Función de transferencia sistema pendulo elástico")


st.latex(r'''
\boxed{F_\text{S}(t) + F_\text{F}(t) + F_\text{I}(t) = F_\text{E}(t)}
''')

st.write("Definiendo cada una de las funciones:")

st.latex(r'''
F_\text{E}(t): \text{ Fuerza externa}
''')

st.latex(r'''
F_\text{S}(t): \text{ Fuerza resorte}
''')

st.latex(r'''
F_\text{F}(t): \text{ Fuerza amortiguador}
''')

st.latex(r'''
F_\text{I}(t): \text{ Fuerza inercia}
''')

st.write("Sus ecuaciones serán:")

st.latex(r'''
F_\text{S}(t) = k y(t)
''')

st.latex(r'''
F_\text{F}(t) = c \frac{d y(t)}{dt}
''')

st.latex(r'''
F_\text{I}(t) = m \frac{d^2 y(t)}{dt^2}
''')

st.write("Reescribiendo la ecuación de acuerdo a las definiciones anteriores:")

st.latex(r'''
m\frac{d^2 y(t)}{dt^2} + c \frac{d y(t)}{dt} + k y(t) = F_\text{E}(t) = x(t)
''')

st.write("Aplicando la transformada de Laplace, se obtiene que:")

st.latex(r'''
ms^2Y(s)+csY(s)+kY(s)=X(s)
''')

st.write("Hallando la función transferencia del sistema masa, resorte y amortiguador:")

st.latex(r'''
\boxed{H(s) = \frac{Y(s)}{X(s)}=\frac{1}{ms^2+cs+k}}
''')

###################

st.write("#### **Función de transferencia circuito RLC serie:**")

st.image("https://lh3.google.com/u/0/d/1R-FOKbYeuORH9E2pI__JeAkA_MTiWK9Q=w1868-h890-iv1")

st.write("Se definen las impedancias complejas y sus equivalentes:")

# Mostrar tabla con imágenes y ecuaciones en LaTeX
st.write(
    """
    | Elemento | $\quad \qquad \qquad \quad \quad$ Dominio del Tiempo $\qquad \qquad \qquad \quad \quad$ | $\qquad \qquad$ Dominio de Laplace $ \qquad \quad$ | Impedancia $Z(s)$ |
    |:---:|:---:|:---:|:---:|
    | ![Resistor](https://github.com/amalvarezme/SenalesSistemas/blob/master/4_Transformada_Laplace/resistor.png?raw=1) | $u(t) = R \cdot i(t)$ | $U(s) = R \cdot I(s)$ | $R$ |
    | ![Inductor](https://github.com/amalvarezme/SenalesSistemas/blob/master/4_Transformada_Laplace/inductor.png?raw=1) | $\\begin{matrix} u(t) = L \\frac{d}{dt} i(t) \\\\ i(t) = \\frac{1}{L} \\int_{0}^{t} u(\\tau) d\\tau + i(0+) \\epsilon(t) \\end{matrix}$ | $\\begin{matrix} U(s) = s L I(s) - L i(0+) \\\\ I(s) = \\frac{1}{s L} U(s) + \\frac{1}{s} i(0+) \\end{matrix}$ | $s L$ |
    | ![Capacitor](https://github.com/amalvarezme/SenalesSistemas/blob/master/4_Transformada_Laplace/capacitor.png?raw=1) | $\\begin{matrix} u(t) = \\frac{1}{C} \\int_{0}^{t} i(\\tau) d\\tau + u(0+) \\epsilon(t) \\\\ i(t) = C \\frac{d}{dt} u(t) \\end{matrix}$ | $\\begin{matrix} U(s) = \\frac{1}{s C} I(s) + \\frac{1}{s} u(0+) \\\\ I(s) = s C U(s) - C u(0+) \\end{matrix}$ | $\\frac{1}{s C}$ |
    """
)

st.write("Resolviendo la malla del circuito y factorizando se tiene que:")

st.latex(r'''
V_i(s) = I(s) [R + \frac{1}{Cs} + Ls]
''')

st.write("Además, para este circuito se tiene que:")

st.latex(r'''
I(s) = Cs \cdot V_c(s)
''')

st.write("Reemplazando $I(s)$ en la ecuación inicial, se obtiene que:")

st.latex(r'''
V_i(s) = Cs \cdot V_c(s) \cdot [R + \frac{1}{Cs} + Ls]
''')

st.write("Finalmente, obteniendo la función de transferencia del circuito RLC serie:")

st.latex(r'''
\boxed{\frac{V_c(s)}{V_i(s)} = \frac{1}{LCs^2 + RCs + 1}}
''')

####################
st.write("#### **Comparación y equivalencia del sistema masa, resorte, amortiguador vs Circuito RLC serie**")

st.write(
    """La comparación de la EDO del circuito RLC seria con la del péndulo elástico
    amortiguado produce equivalencia entre sistemas, por lo tanto se puede relacionar
    estas equivalencias mediante la siguiente tabla
    """
)

st.write(
    """
|                         | Circuito RLC | Péndulo elástico                          |
|-------------------------|------------------------------------|-------------------------------------------|
| Señal de entrada $x(t)$ | $V_\text{i}(t) = F_\text{E}(t)$  | $F_\text{E}(t) = V_\text{i}(t)$ |
| Señal de salida $y(t)$  | $V_\text{c}(t)$                    | $y(t)$                                    |
|                         | $LC=m$                              | $m=LC$                                     |
|                         | $RC=c$                              | $c=RC$                                     |
|                         | $1=k$                    | $k=1$                           | |
    """
)

#################
st.write("#### **Simulación diferentes tipos de amortiguamiento**")

import numpy as np
import matplotlib.pyplot as plt

import sympy as sym
sym.init_printing()
#definiri variables
t, m, c, k = sym.symbols('t m c k', real=True,positive=True)
s = sym.symbols('s', complex=True)

#forma canónica del slit
a2 = m #LC en circuito equivalente
a1 = c #RC en circuito equivalente
ao = k #1 en circuito equivalente

Ko = 1/ao #ganancia circuito forma canónica
wn = sym.sqrt(ao/a2)
xi = a1/(2*sym.sqrt(ao*a2))

c_val = 1.0  # Fijo
k_val = 1.0  # Fijo

#Solicitar al usuario que ingrese el valor de xi
st.write("###### **Ingrese el valor del factor de amortiguamiento, valores entre 0 y 1 sera un sistema subamortiguado, valor de 1 criticamente amortiguado, valores mayores a 1 sobreamortiguado**")
xi_val = st.number_input("Introduce el valor del factor de amortiguamiento: ", value= 0.25)

#Usar la ecuación despejada para m
m_val = c_val**2 / (4 * xi_val**2 * k_val)

st.write("##### **Función de transferencia**")
#función de transferencia
H = Ko*wn**2/(s**2+2*xi_val*wn*s+wn**2)
H

st.write("##### **Polos de la función de transferencia**")
#calcular polos
p1 = -xi_val*wn + 1j*wn*sym.sqrt(1-xi_val**2)
p2 = -xi_val*wn - 1j*wn*sym.sqrt(1-xi_val**2)
p1, p2

st.write("##### **Se fijan los valores para el péndulo**")
#Fijando los valores para el pendulo
mck = {m: m_val, c: c_val, k: k_val}
st.write("###### Valores de mck: ")
st.write("Masa: ", m_val)
st.write("Amortiguador: ", c_val)
st.write("Resorte: ", k_val)

st.write("#### **Cálculo de R, L, C a partir de los valores de m, k y c**")
#Como k = 1, se deduce que C = 1
C_val = 1

#Calcular L y R usando las ecuaciones dadas y despejadas
L_val = m_val / C_val
R_val = c_val / C_val

#Mostrar los resultados
st.write("El valor de L es: ", L_val)
st.write("El valor de R es: ", R_val)
st.write("El valor de C es: ", C_val)

st.write("#### **Gráficas de polos y ceros**")
#gráfica polos y ceros paquete de control de sympy
from sympy.physics.control.lti import TransferFunction
from sympy.physics.control.control_plots import pole_zero_plot, bode_plot, impulse_response_plot, step_response_plot
num,den=sym.fraction(H.subs(mck))

tf1 = TransferFunction(num, den, s)#numerados y denomirador función de transferencia con valores evaluados de RLC
fig, ax = plt.subplots()
pole_zero_plot(tf1, ax=ax)

st.pyplot(fig)

#Diagrama de Bode
st.write("#### **Diagrama de Bode**")
fig, ax = plt.subplots()
bode_plot(tf1)

st.pyplot(fig)

#Respuesta al impulso
st.write("#### **Respuesta al impulso**")
fig, ax = plt.subplots()
impulse_response_plot(tf1,upper_limit=5/(xi_val*wn.subs(mck)))

st.pyplot(fig)

#Respuesta al escalón
st.write("#### **Respuesta al escalón**")
fig, ax = plt.subplots()
step_response_plot(tf1,upper_limit=5/(xi_val*wn.subs(mck)))

st.pyplot(fig)

#Caso Oscilatorio
st.write("## **Caso oscilatorio**")

#forma canónica del slit
a21 = m #LC en circuito equivalente
a11 = c #RC en circuito equivalente
ao1 = k #1 en circuito equivalente

Ko1 = 1/ao1 #ganancia circuito forma canónica
wn1 = sym.sqrt(ao1/a21)
xi = 0

st.write("##### **Función de transferencia para el caso en el que XI = 0**")
#función de transferencia
H2 = Ko1*wn1**2/(s**2+2*xi*wn1*s+wn1**2)
H2

st.write("##### **Polos de la función de transferencia**")
#calcular polos
p11 = -xi*wn1 + 1j*wn1*sym.sqrt(1-xi**2)
p21 = -xi*wn1 - 1j*wn1*sym.sqrt(1-xi**2)
p11, p21

#Fijando los valores para el pendulo
mck1 = {m: 4, c: 1, k: 1}

st.write("#### **Gráficas de polos y ceros**")
#gráfica polos y ceros paquete de control de sympy
from sympy.physics.control.lti import TransferFunction
from sympy.physics.control.control_plots import pole_zero_plot, bode_plot, impulse_response_plot, step_response_plot
num1,den1=sym.fraction(H2.subs(mck1))

tf2 = TransferFunction(num1, den1, s)#numerados y denomirador función de transferencia con valores evaluados de RLC
fig, ax = plt.subplots()
pole_zero_plot(tf2, ax=ax)

st.pyplot(fig)

#Diagrama de Bode
st.write("#### **Diagrama de Bode**")
fig, ax = plt.subplots()
bode_plot(tf2)

st.pyplot(fig)

#Respuesta al impulso
st.write("#### **Respuesta al impulso**")
fig, ax = plt.subplots()
impulse_response_plot(tf2,upper_limit=5/(xi_val*wn.subs(mck)))

st.pyplot(fig)

#Respuesta al escalón
st.write("#### **Respuesta al escalón**")
fig, ax = plt.subplots()
step_response_plot(tf2,upper_limit=5/(xi_val*wn.subs(mck)))

st.pyplot(fig)
```

```python
%%writefile 3_📻_Filtrado_y_detección_de_genero.py

import streamlit as st
import matplotlib.pyplot as plt
import numpy as np
import soundfile as sf # para instalar pip install soundfile
import subprocess
import os

st.set_page_config(page_title="Filtrado y detección de genero", page_icon="📻")

st.markdown("# Filtrado y detección de genero")
st.sidebar.header("Filtrado y detección de genero")
st.write(
    """Este punto muestra el procedimiento para la modulación en amplitud (AM) de un mensaje, en este
    caso cualquier canción de YouTube en la cual solo se procesan 5 segundos (desde el segundo 20 al 25),
    el usuario puede ingresar cualquier valor para el indice de modulación, asi como ingresar cualquier link de
    YouTube para procesar, además se realiza tambien el proceso de demodulación y filtrado de la señal haciendo uso
    de un filtro pasabajas Butterworth, una vez realizado este proceso de modulación, demodulación y filtrado se
    procede a detectar el genero de la canción."""
)



st.write("Demodulador")
st.image("https://github.com/amalvarezme/SenalesSistemas/blob/master/Otros/AM.jpeg?raw=1")

# Función para descargar el audio usando yt-dlp
def download_audio(link):
    try:
        # Ejecutar el comando yt-dlp para descargar el audio
        result = subprocess.run(
            ["yt-dlp", "--extract-audio", "--audio-format", "mp3", "-o", "audio.mp3", "--force-overwrites", link],
            capture_output=True, text=True
        )
        if result.returncode == 0:
            st.success("Audio descargado exitosamente.")
        else:
            st.error(f"Error al descargar el audio: {result.stderr}")
    except Exception as e:
        st.error(f"Error al ejecutar el comando: {e}")

# Función para convertir MP3 a WAV usando ffmpeg
def convert_to_wav():
    try:
        # Ejecutar el comando ffmpeg para convertir MP3 a WAV
        result = subprocess.run(
            ["ffmpeg", "-y", "-i", "audio.mp3", "output.wav"],
            capture_output=True, text=True
        )
        if result.returncode == 0:
            st.success("Conversión a WAV completada.")
        else:
            st.error(f"Error al convertir el archivo: {result.stderr}")
    except Exception as e:
        st.error(f"Error al ejecutar el comando: {e}")



# Interfaz en Streamlit
st.markdown("# Procesar Audio de YouTube")

# Enlace de YouTube ingresado por el usuario
link = st.text_input("Introduce el enlace de YouTube:", "https://www.youtube.com/watch?v=vq-zaD1poAE")

# Si el enlace es proporcionado, ejecutar las funciones
if link:
    download_audio(link)
    convert_to_wav()


import numpy as np
import soundfile as sf

#lee archivos wav
nombre_out = "output.wav"
x, fs = sf.read(nombre_out)

if x.ndim > 1:
    x = np.mean(x, axis=1)

# Extraer los 5 segundos desde los 20 a los 25 segundos
start_sample = int(20 * fs)
end_sample = int(25 * fs)
x_fragment = x[start_sample:end_sample]

#Reproducir el fragmento seleccionado
with open('output.wav', 'rb') as audio_file:
    st.audio(audio_file, format='audio/wav', start_time=20, end_time=25)

import matplotlib.pyplot as plt

st.markdown("## Señal mensaje")

st.markdown("### Señal mensaje en el tiempo")
import matplotlib.pyplot as plt
import numpy as np
import soundfile as sf # para instalar pip install soundfile
from IPython.display import Audio #librería para reproducir audio

#leer archivo wav
nombre_out = "output.wav"
xa, fs = sf.read(nombre_out)

# read speech signal from file
print('Frecuencia de muestreo %.2f[Hz]\naudio %s' % (fs,nombre_out))

xpro = xa.copy() #copiar archivos para procesar
#leer 5 segundos de la canción
ti = 20 #tiempo incio a procesar seg
tf = 25 #tiempo final a procesar seg
m = xpro[int(ti*fs):int((tf*fs)),0] #señal mensaje en primer canal

tt = np.arange(ti,tf,1/fs) # vector de tiempo

fig, ax = plt.subplots()
ax.plot(tt, m)
ax.set_xlabel('$tiempo [s]$')
ax.set_ylabel('$m(t)$')

st.pyplot(fig)

st.markdown("### Señal mensaje en frecuencia (Magnitud y fase)")
# Primera gráfica: Magnitud del espectro
Mw = np.fft.rfft(m) # axis=0 permite aplicar fft por cada columna de xpro
vf = np.fft.rfftfreq(m.shape[0],1/fs) #se crea el vector de frecuencias
fig1, ax1 = plt.subplots()
ax1.plot(vf, abs(Mw))
ax1.set_title(r'Espectro señal mensaje')
ax1.set_xlabel(r'$f[Hz]$', fontsize=14)
ax1.set_ylabel(r'$|M[f]|$', fontsize=14)

# Mostrar la primera gráfica en Streamlit
st.pyplot(fig1)

# Segunda gráfica: Fase del espectro
fig2, ax2 = plt.subplots()
ax2.plot(vf, np.angle(Mw))
ax2.set_xlabel(r'$f[Hz]$', fontsize=14)
ax2.set_ylabel(r'$<M[f]$', fontsize=14)

# Mostrar la segunda gráfica en Streamlit
st.pyplot(fig2)


st.markdown("## Señal portadora")
#parámetros de la portadora
Fc = 15000  #frecuencia de la portadora
Im = st.number_input("Ingrese el índice de modulación: ", 0.8) #Im = peak(m(t))/Ac ->  Ac = peak(m(t))/Im
Fs = 5 * Fc

Ac = np.max(np.abs(m)) / Im #amplitud de la portadora según índice de modulación
c = Ac * np.sin(2 * np.pi * Fc * tt)  #portadora

st.write("Audio señal portadora")
#repoducir portadora
st.audio(c,sample_rate=fs)

st.markdown("### Señal portadora en el tiempo")
# Crear la gráfica de la señal portadora
fig, ax = plt.subplots()
ax.plot(tt, c)
ax.set_xlabel('$tiempo [s]$')
ax.set_ylabel('$c(t)$')

# Mostrar la gráfica en Streamlit
st.pyplot(fig)

st.markdown("### Señal portadora en frecuencia (Magnitud y fase)")
#Fourier señal portadora
Cw = np.fft.rfft(c) # axis=0 permite aplicar fft por cada columna de xpro
# Gráfica de la magnitud del espectro de la portadora
fig1, ax1 = plt.subplots()
ax1.plot(vf, abs(Cw))
ax1.set_title(r'Espectro Portadora')
ax1.set_xlabel(r'$f[Hz]$', fontsize=14)
ax1.set_ylabel(r'$|C[f]|$', fontsize=14)
st.pyplot(fig1)

# Gráfica de la fase del espectro de la portadora
fig2, ax2 = plt.subplots()
ax2.plot(vf, np.angle(Cw))
ax2.set_xlabel(r'$f[Hz]$', fontsize=14)
ax2.set_ylabel(r'$<C[f]$', fontsize=14)
st.pyplot(fig2)


st.markdown("## Señal modulada")
#señal modulada
A1 = 2
y = A1*m*c #señal modulada en AM

st.write("Audio señal modulada")
st.audio(y,sample_rate=fs)

st.markdown("### Señal modulada en el tiempo")
# Gráfica de la señal modulada
fig, ax = plt.subplots()
ax.plot(tt, y)
ax.set_xlabel('$tiempo [s]$')
ax.set_ylabel('$y(t)$')

# Mostrar la gráfica en Streamlit
st.pyplot(fig)

st.markdown("### Señal modulada en frecuencia (Magnitud y fase)")
#Fourier señal modulada
Yw = np.fft.rfft(y) # axis=0 permite aplicar fft por cada columna de xpro
# Gráfica de la magnitud del espectro de la señal modulada
fig1, ax1 = plt.subplots()
ax1.plot(vf, abs(Yw))
ax1.set_title(r'Espectro modulada')
ax1.set_xlabel(r'$f[Hz]$', fontsize=14)
ax1.set_ylabel(r'$|Y[f]|$', fontsize=14)
st.pyplot(fig1)

# Gráfica de la fase del espectro de la señal modulada
fig2, ax2 = plt.subplots()
ax2.plot(vf, np.angle(Yw))
ax2.set_xlabel(r'$f[Hz]$', fontsize=14)
ax2.set_ylabel(r'$<Y[f]$', fontsize=14)
st.pyplot(fig2)


st.markdown("## Señal demodulada")
d = y*c #señal demodulada

st.write("Audio señal demodulada")
st.audio(d,sample_rate=fs)

st.markdown("### Señal demodulada en el tiempo")
# Gráfica de la señal demodulada
fig, ax = plt.subplots()
ax.plot(tt, d)
ax.set_xlabel('$tiempo [s]$')
ax.set_ylabel('$d(t)$')

# Mostrar la gráfica en Streamlit
st.pyplot(fig)


st.markdown("### Señal demodulada en frecuencia (Magnitud y fase)")
#Fourier señal demodulada
Dw = np.fft.rfft(d) # axis=0 permite aplicar fft por cada columna de xpro
# Gráfica de la magnitud del espectro de la señal demodulada
fig1, ax1 = plt.subplots()
ax1.plot(vf, abs(Dw))
ax1.set_title(r'Espectro demodulada')
ax1.set_xlabel(r'$f[Hz]$', fontsize=14)
ax1.set_ylabel(r'$|D[f]|$', fontsize=14)
st.pyplot(fig1)

# Gráfica de la fase del espectro de la señal demodulada
fig2, ax2 = plt.subplots()
ax2.plot(vf, np.angle(Dw))
ax2.set_xlabel(r'$f[Hz]$', fontsize=14)
ax2.set_ylabel(r'$<D[f]$', fontsize=14)
st.pyplot(fig2)

###############
st.markdown("## Diseño filtro digital pasabajas")
from scipy.signal import butter as bw
from scipy.signal import freqz_zpk

N = 15 #orden el filtro
Wn = 14950 #banda pasante
filt = 'lowpass' #tipo del filtro (pasa-baja, pasa-alta, pasabanda, rechazabanda)
out = 'zpk' #la funcion retorna los polos y ceros junto con la ganancia. Para usar numerador y denominador usar 'ba' junto con freqz
zeros, poles, gain = bw(N, Wn, btype=filt, output=out, fs=fs)
w, h = freqz_zpk(zeros, poles, gain, fs=fs) #Para indicar la longitud usar worN=K

#Definicion de funciones para las graficas
from matplotlib.patches import Circle
#gráfica polos y ceros
def plot_freq_response(filter_name, w, h, N): #Mostrar la respuesta en frecuencia del filtro
  fig22 = plt.figure()

  ax1 = fig.add_subplot(1, 1, 1)
  ax1.set_title(filter_name +' Digital filter frequency response, order= ' + str(N))
  ax1.plot(w, 20 * np.log10(abs(h)), 'b')
  ax1.set_ylabel('Amplitude [dB]', color='b')
  ax1.set_xlabel('Frequency [Hz]')
  ax1.grid()

  ax2 = ax1.twinx()
  angles = np.unwrap(np.angle(h))
  ax2.plot(w, angles, 'g')
  ax2.set_ylabel('Angle [radians]', color='g')
  plt.axis('tight')
  plt.show()

def show_zp(z, p, title= 'Z-plane'): # Mostrar la ubicación de los polos y los zeros
  fig20, ax = plt.subplots()

  ax.plot(np.real(z), np.imag(z), 'bo', fillstyle='none', ms = 10)
  ax.plot(np.real(p), np.imag(p), 'rx', fillstyle='none', ms = 10)
  unit_circle = Circle((0,0), radius=1, fill=False,
                        color='black', ls='solid', alpha=0.9)
  ax.add_patch(unit_circle)
  ax.axvline(0, color='0.7')
  ax.axhline(0, color='0.7')

  plt.title(title)
  plt.xlabel(r'Re{$z$}')
  plt.ylabel(r'Im{$z$}')
  plt.axis('equal')
  plt.xlim((-2, 2))
  plt.ylim((-2, 2))
  plt.grid()

  st.pyplot(fig20)

st.markdown("### Gráfica de polos y ceros del filtro")
show_zp(zeros, poles, title='Polos y Ceros')#gráfica polos y ceros

st.markdown("### Diagrama de Bode del filtro pasabajas")
st.image("https://lh3.google.com/u/0/d/1ALFQr2kPgJoOAGceS_LRrKgPBX4jjjCp=w1868-h890-iv1")
plot_freq_response('Butterworth', w, h, N) #gráfica de bode del filtro


st.markdown("## Audio filtrado")
#filtrar señal
from scipy.signal import filtfilt

num,den = bw(N, Wn, btype=filt, fs=fs)

xf = filtfilt(num, den, d, axis=0)#filtro pasabajos sobre señal demodulada d
me = (2/A1)*xf #señal reescalada

st.write("Audio señal filtrada")
st.audio(me,sample_rate=fs)

st.markdown("### Señal filtrada en el tiempo")
# Gráfica de la señal recuperada
fig, ax = plt.subplots()
ax.plot(tt, me)
ax.set_xlabel('$tiempo [s]$')
ax.set_ylabel('$m_e(t)$')

# Mostrar la gráfica en Streamlit
st.pyplot(fig)


st.markdown("### Señal filtrada en frecuencia (Magnitud y fase)")
#Fourier señal recuperada
Mew = np.fft.rfft(me) # axis=0 permite aplicar fft por cada columna de xpro
# Gráfica de la magnitud del espectro de la señal recuperada
fig1, ax1 = plt.subplots()
ax1.plot(vf, abs(Mew))
ax1.set_title(r'Espectro recuperada')
ax1.set_xlabel(r'$f[Hz]$', fontsize=14)
ax1.set_ylabel(r'$|M_e[f]|$', fontsize=14)
st.pyplot(fig1)

# Gráfica de la fase del espectro de la señal recuperada
fig2, ax2 = plt.subplots()
ax2.plot(vf, np.angle(Mew))
ax2.set_xlabel(r'$f[Hz]$', fontsize=14)
ax2.set_ylabel(r'$<M_e[f]$', fontsize=14)
st.pyplot(fig2)

################ Deteccion de genero

import os
import streamlit as st
import joblib
import numpy as np
import gdown

# Crear la carpeta "modelo" si no existe
if not os.path.exists('modelo'):
    os.makedirs('modelo')

# Función para descargar el modelo desde Google Drive
def descargar_modelo(drive_id):
    url = f'https://drive.google.com/uc?id={drive_id}'
    output = 'modelo/reggaeton_vs_metal.pkl'
    gdown.download(url, output, quiet=False)

# ID del archivo en Google Drive
drive_id = '1suOtvgobMdPwMfZASdkiatZWksPpRSFs'

# Ruta del modelo
modelo_path = 'modelo/reggaeton_vs_metal.pkl'

# Descargar el modelo si no está disponible
if not os.path.exists(modelo_path):
    descargar_modelo(drive_id)

# Cargar el modelo
loaded_model = joblib.load(modelo_path)

# Verificación del cargue del modelo
st.write("Modelo cargado con éxito.")
st.write(loaded_model.keys())

def normalizar_espectro(espectro):
    return espectro / np.linalg.norm(espectro)

def detectar_genero(me, modelo):
    #cargar variables del modelo
    Xw_ = modelo['Xw_']  #base de datos de espectros
    label = modelo['label']  #géneros de las canciones

    #calcular la rfft del nuevo registro (Xfn)
    Xfn = np.abs(np.fft.rfft(me))

    #verificar dimensiones
    print(f"Dimensiones de Xfn (nueva canción): {Xfn.shape}")
    print(f"Dimensiones de Xw_ (base de datos): {Xw_.shape}")

    #normalizar espectros
    Xfn_normalizado = normalizar_espectro(Xfn)
    Xw_normalizado = np.array([normalizar_espectro(x) for x in Xw_])

    #calcular la distancia usando los espectros normalizados
    distancias = np.linalg.norm(Xw_normalizado - Xfn_normalizado, axis=1)

    # Mostrar distancias
    #print(f"Distancias calculadas: {distancias}")

    #encontrar el índice de la distancia mínima
    indice_minimo = np.argmin(distancias)

    #detectar el género correspondiente (1: Metal, 2: Reggaetón)
    genero_detectado = label[indice_minimo]
    cancion_parecida = modelo['name_c'][indice_minimo]

    return genero_detectado, cancion_parecida

#ejecutar la función de detección
genero, cancion_parecida = detectar_genero(me, loaded_model)

st.markdown("## Genero detectado de la canción demodulada y filtrada")

#mostrar el género detectado y la canción más parecida
if genero == 1:
    st.write(f"El género detectado es: Metal, Canción más parecida: {cancion_parecida}")
elif genero == 2:
    st.write(f"El género detectado es: Reggaetón, Canción más parecida: {cancion_parecida}")
```
